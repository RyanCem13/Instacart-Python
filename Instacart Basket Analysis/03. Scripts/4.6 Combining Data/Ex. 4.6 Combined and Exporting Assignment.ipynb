1. Import Libraries, Create Path and Import Datasets

```python
#Importing Libraries
import pandas as pd
import numpy as nb
import os
```

```python
#Create Path
path = r'C:\Users\ryanc\OneDrive\Documents\Data Analytics Exercises\Python\Instacart Basket Analysis'
```

```python
#Import order products combined df
df_combined = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'orders_products_combined.pkl'))
```

2. Explore and Prepare dataframes to merge

```python
df_combined.shape
```

```python
#Import products df
df_prod = pd.read_csv(os.path.join(path, '02 Data', 'Prepared Data', 'products_checked.csv'), index_col = False)
```

```python
#View Combined Data
df_combined.head()
```

```python
#View Product Data
df_prod.head()
```

```python
#Drop Merge Column from combined
df_combined = df_combined.drop(columns = '_merge')
```

```python
#Check column is dropped
df_combined.head()
```

```python
#Drop Unnamed: 0 column from products
df_prod = df_prod.drop(columns = 'Unnamed: 0')
```

```python
#Check Column is dropped
df_prod.head()
```

3. Merge Dataframes

```python
#Combine df's with indicator
df_merged_all = df_prod.merge(df_combined, on = 'product_id', indicator = True)
```

```python
#Check Merged Data
df_merged_all.head()
```

```python
#Check Merge Flag
df_merged_all['_merge'].value_counts()
```

4. Export merged data to pickle

```python
#Export as pickle
df_merged_all.to_pickle(os.path.join(path, '02 Data','Prepared Data', 'orders_products_merged.pkl'))
```
