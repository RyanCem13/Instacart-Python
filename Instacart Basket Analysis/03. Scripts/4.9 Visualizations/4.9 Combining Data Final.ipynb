1. Import Libraries, Create Path and Import Datasets

```python
#import libraries
import pandas as pd
import numpy as np
import os
```

```python
#Create Path
path = r'C:\Users\ryanc\OneDrive\Documents\Data Analytics Exercises\Python\Instacart Basket Analysis'
```

```python
#Import order products merged updated df
ords_prods_merge = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'orders_products_merged_updated.pkl'))
```

```python
#Import customer df
customers = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'customers_updated_final.pkl'))
```

2. Prepare df's to merge

```python
#View ords_prods_merge head
ords_prods_merge.head()
```

```python
#View customer df head
customers.head()
```

Column in common  is user_id

```python
#Check dtype
customers['user_id'].dtype
```

```python
#Check dtypes
ords_prods_merge['user_id'].dtype
```

```python
#Change dtype
customers['user_id']=customers['user_id'].astype('int32')
```

```python
#Check dtype
customers['user_id'].dtype
```

3. Combine order_prods_merge and Customers df's

```python
#Combine df's with indicator
final_merged = ords_prods_merge.merge(customers, on = 'user_id', indicator = True)
```

```python
#View all columns setting change
pd.set_option('display.max_columns', None)
```

```python
#View final_merged data
final_merged.head(25)
```

4. Export new merged dataset

```python
#Export final_merge data
final_merged.to_pickle(os.path.join(path, '02 Data','Prepared Data', 'final_merged.pkl'))
```
